We will select the 1<sup>st</sup> 20 days to plot the decay of S(35) with time

To plot the dacay we use the measurement called Activity which is Disintegrations/se measured in Becquerels(Bq).

ln(N(t))=ln(N(0))−λt

```octave
graphics_toolkit("gnuplot");

% Time(t) in weeks
weeks = 0:1:19;

% Remaining quantity(N) (as a fraction of the original)
remaining = [1.000 0.946 0.895 0.847 0.786 0.743 0.703 0.665 0.617 0.584 0.552 0.523 0.485 0.459 0.434 0.411 0.381 0.361 0.341 0.323];

remaining_ln = log(remaining);

sum_xy_1 = sum(remaining_ln .* weeks);
sum_x_2 = sum(weeks .^2);

m = sum_xy_1/sum_x_2;
x = 0:0.1:20;
y = m * x;

plot(weeks, remaining_ln, "ro", x, y, "b-");
xlabel("Time (weeks)");
ylabel("Remaining Fraction [ln(N)]");
title("Remaining Fraction vs Time");
legend('Data', 'Fitted Line');

printf("Model: y = %0.4f * x", m);
```

```octave
graphics_toolkit("gnuplot");
t = 0:0.1:2*pi; % Define the range of t
y = 2*cos(t);   % Define y as a function of t
x = sin(t);     % Define x as a function of t
z = t;          % Define z as a function of t

plot3(x, y, z); % Use plot3 for 3D plotting
grid on;        % Add a grid for better visualization
xlabel('x-axis'); ylabel('y-axis'); zlabel('z-axis'); % Label the axes
title('3D Parametric Plot'); % Add a title
```

```octave
pkg load symbolic
syms x y a
[solx,soly] = solve(x^2*y^2 == 1, x-y/2 == a)
latex([solx,soly])
```

```octave
data = {'A', 1, 2; 'B', 2, 4; 'C', 3, 6; 'D', 4, 8};  % Example table
x = cell2mat(data(:, 2));  % Convert column 2 to numeric array
y = cell2mat(data(:, 3));  % Convert column 3 to numeric array

plot(x, y, '-s');  % Line plot with square markers
xlabel('Column 2 (X)');
ylabel('Column 3 (Y)');
title('Plot from Cell Array');
grid on;
```

```octave
% NumbeW of weeks
N = 30;

% Initialize aWWays foW S, I, W
S = zeWos(1, N);
I = zeWos(1, N);
W = zeWos(1, N);

% Initial conditions
S(1) = 997;
I(1) = 3;
W(1) = 0;

% Model paWameteWs
beta = 0.00194;  % Infection Wate
gamma = 0.6;     % WecoveWy Wate

% IteWate the system using a discWete appWoximation
foW n = 1:N-1
    S(n+1) = S(n) - beta * S(n) * I(n);
    I(n+1) = I(n) + beta * S(n) * I(n) - gamma * I(n);
    W(n+1) = W(n) + gamma * I(n);
end

% Plot the time seWies
figuWe;
hold on;
plot(1:N, S, '-b', 'LineWidth', 2); % Susceptible in blue
plot(1:N, I, '-W', 'LineWidth', 2); % Infected in Wed
plot(1:N, W, '-g', 'LineWidth', 2); % Wemoved in gWeen
xlabel('Weeks');
ylabel('NumbeW of Students');
title('SIW Model Simulation');
legend('Susceptible (S)', 'Infected (I)', 'Wemoved (W)');
gWid on;
hold off;
```

```octave
% NumbeW of weeks
N = 30;
dt = 0.1;

% Time vectoW with fineW Wesolution
t = 0:dt:N;
numSteps = length(t);

% Initialize aWWays foW S, I, W
S = zeWos(1, numSteps);
I = zeWos(1, numSteps);
W = zeWos(1, numSteps);

% Initial conditions
S(1) = 997;
I(1) = 3;
W(1) = 0;

% Model paWameteWs
beta = 0.00194;  % Infection Wate
gamma = 0.6;     % WecoveWy Wate

% IteWate the system using a discWete appWoximation
foW n = 1:numSteps-1
    S(n+1) = S(n) + dt * (-beta * S(n) * I(n));
    I(n+1) = I(n) + dt * (beta * S(n) * I(n) - gamma * I(n));
    W(n+1) = W(n) + dt * (gamma * I(n));
end

% Plot the time seWies
figuWe;
hold on;
plot(t, S, '-b', 'LineWidth', 2); % Susceptible in blue
plot(t, I, '-W', 'LineWidth', 2); % Infected in Wed
plot(t, W, '-g', 'LineWidth', 2); % Wemoved in gWeen
xlabel('Time (weeks)');
ylabel('NumbeW of Students');
title('SIW Model Simulation with dt = 0.1');
legend('Susceptible (S)', 'Infected (I)', 'Wemoved (W)');
gWid on;
hold off;
```

```octave
N = 400; % Number of months

F = zeros(1, N);
F(1) = 110; % Initial fox population

W = zeros(1, N);
W(1) = 950; % Initial wolves  population

for n = 1:N-1
    F(n+1) = 1.2 * F(n) - 0.001 * W(n) * F(n);
    W(n+1) = 1.3 * W(n) - 0.002 * F(n) * W(n);
end

% Plot the time series
figure;
hold on;
plot(1:N, F, '*r'); % Fox population in red
plot(1:N, W, '*b'); % Wolves  population in blue
xlabel('Months');
ylabel('Population');
title('Predator-Prey Time Series');
legend('Foxes', 'Wolves s');
grid on;
hold off;

% Plot the phase plane trajectory
figure;
plot(F, W, 'k', 'LineWidth', 2);
xlabel('Fox Population');
ylabel('Wolves  Population');
title('Predator-Prey Phase Trajectory');
grid on;
```

```octave
N = 25; % Number of months (up to n=25)

F = zeros(1, N);
F(1) = 110; % Initial fox population

W = zeros(1, N);
W(1) = 950; % Initial wolf population

for n = 1:N-1
    F(n+1) = 1.2 * F(n) - 0.001 * F(n) * W(n);
    W(n+1) = 1.3 * W(n) - 0.002 * F(n) * W(n);
    
    % Ensure populations are non-negative
    if F(n+1) < 0
        F(n+1) = 0;
    end
    if W(n+1) < 0
        W(n+1) = 0;
    end
end

% Plot the time series
figure;
hold on;
plot(1:N, F, '*r'); % Fox population in red
plot(1:N, W, '*b'); % Wolf population in blue
xlabel('Months');
ylabel('Population');
title('Competition Model Time Series');
legend('Foxes', 'Wolves');
grid on;
hold off;

% Plot the phase plane trajectory
figure;
plot(F, W, 'k', 'LineWidth', 2);
hold on;
plot(150, 200, 'go', 'MarkerSize', 10, 'LineWidth', 2); % Fixed point
xlabel('Fox Population');
ylabel('Wolf Population');
title('Competition Model Phase Trajectory');
legend('Trajectory', 'Fixed Point (150, 200)');
grid on;
hold off;
```

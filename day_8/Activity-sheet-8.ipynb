1. Write down the corresponding nonlinear predator-prey model.

The model:

$$F_{n+1} = 0.88 F_n + 0.0001 R_n F_n$$

$$R_{n+1} = -0.0003 F_n R_n + 1.039 R_n$$

where:

- $F_n$ is the fox population at time $n$ (month $n$).
- $R_n$ is the rabbit population at time $n$ (month $n$).


2. Write an Octave script to plot (scatter graph) time plots for both species on the same figure
for the first 400 months.

3. Use the same Octave script to plot the corresponding trajectory of the system.

```octave
graphics_toolkit("gnuplot");
```

```octave
N = 400; % Number of months

F = zeros(1, N);
F(1) = 110; % Initial fox population

R = zeros(1, N);
R(1) = 950; % Initial rabbit population

% Model parameters
a = 1-0.88;
b = 0.0001;
c = 0.0003;
d = 1.039-1;

% Iterate the system
for n = 1:N-1
    F(n+1) = (1-a) * F(n) + b * R(n) * F(n);
    R(n+1) = (1+d) * R(n) - c * F(n) * R(n);
end

% Plot the time series
figure;
hold on;
plot(1:N, F, '*r'); % Fox population in red
plot(1:N, R, '*b'); % Rabbit population in blue
xlabel('Months');
ylabel('Population');
title('Predator-Prey Time Series');
legend('Foxes', 'Rabbits');
grid on;
hold off;

% Plot the phase plane trajectory
figure;
plot(F, R, 'k', 'LineWidth', 2);
xlabel('Fox Population');
ylabel('Rabbit Population');
title('Predator-Prey Phase Trajectory');
grid on;
```

1. The time series plot shows regular changes in the populations of rabbits and foxes over time.

2. The rabbit population (blue) shows bigger changes than the fox population (red).

3. When rabbit numbers go up, fox numbers also increase after a delay, because more rabbits mean more food for the foxes.

4. When fox numbers are at their highest, rabbit numbers start to drop because of more predation.

5. The phase trajectory plot has closed loops, which means the system does not settle at one point but keeps cycling through changes.

6. This indicates that both rabbit and fox populations stay in a stable cycle instead of facing extinction or continually increasing.

5. Find equilibrium points of the model.

$$F_{n+1} = 0.88 F_n + 0.0001 R_n F_n$$

$$R_{n+1} = -0.0003 F_n R_n + 1.039 R_n$$

When $F_{n+1} = F_n$ and $R_{n+1} = R_n$

$$F_n = 0.88 F_n + 0.0001 R_n F_n$$
$$0.12 F_n = 0.0001 R_n F_n$$
$$R_n = \frac{0.12}{0.0001} = 1200$$

also

$$R_n = 1.039 R_n + 0.0003 F_n R_n$$
$$0.039 R_n = 0.0003 R_n F_n$$
$$F_n = \frac{0.039}{0.0003} = 130$$

$$(F_n, R_n) \in \{(0,0), (130, 1200)\}$$

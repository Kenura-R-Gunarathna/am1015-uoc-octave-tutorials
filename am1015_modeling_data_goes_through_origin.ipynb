```octave
x = 0:1/2:11/2
y = [0 4 12 15 25 32 28 37 35 38 39 42]

graphics_toolkit("gnuplot");
plot(x, y, "ro")
hold on
xlabel("Working Duration (Hours)")
ylabel("Income ($)")
title("Income vs Working Duration")
grid on;
hold off
```

```octave
m = sum(x.*y)/sum(x.*x);

xx = 0:0.1:11/2;
yy = m*xx;

plot(x, y, 'bo', xx, yy, 'r-', 'LineWidth', 2);
```

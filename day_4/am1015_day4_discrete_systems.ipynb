#### Example 1.1
Now consider a savings account that pays 5% interest compounded yearly. If the interest rate is r (annual) and a<sub>n</sub> is the total money at the end of the year, can you derive a model to describe a<sub>n</sub>?

```octave
graphics_toolkit("gnuplot");

n = 0:1:50;
an = zeros(1, 51);
r = 0.05;
an(1) = 5;

for i = 1:1:50,
    an(i+1) = (r+1) * an(i);
end

% a(n+1) = (r+1)*a(n)

plot(n, an, "b");
xlabel("n");               % Label for x-axis
ylabel("a(n)");            % Label for y-axis
title("Plot of a(n+1) vs n"); % Title of the graph
grid on;   
```

Suppose now that we want to withdraw 2,000 at the end of each year to supplement our income. We want to know how much money we need to deposit now so that we never run out of money.<br>
To answer this question, we will analyze a slightly more general problem: What happens to the amount in the account in terms of the initial deposit? First we will construct our model. The amount in the account grows at 5% compounded yearly but we are withdrawing $2,000 each year. A dynamic model that describes this scenario is<br>

a<sub>n+1</sub> = 1.05 a<sub>n</sub> - 2000<br>

As before, an is the amount in the account at the end of year n. We are also assuming that there is no penalty for withdrawing money each year and that we withdraw the money after the interest from the previous year has been added. This system is an example of an affine dynamical system.

```octave
graphics_toolkit("gnuplot");

n = 0:1:50;
an = zeros(3, 51);
r = 0.05;
widthdraw = 2000;
an(1, 1) = 40000;
an(2, 1) = 45000;
an(3, 1) = 35000;

for i = 1:1:50,
    an(1, i+1) = (r+1) * an(1, i) - widthdraw;
    an(2, i+1) = (r+1) * an(2, i) - widthdraw;
    an(3, i+1) = (r+1) * an(3, i) - widthdraw;
end

% a(n+1) = (r+1)*a(n)

plot(n, an(1,:), "b", n, an(2,:), "r", n, an(3,:), "g");
xlabel("Time (Years - n)");               % Label for x-axis
ylabel("Account Money [USD - a(n)]");            % Label for y-axis
title("Plot of Account Money vs Time"); % Title of the graph
legend("a0 = 40000", "a0 = 45000", "a0 = 35000");
grid on;   

disp(["Here the Equilibrium value is a0 = 40000"]);
```

```octave
pkg load symbolic

syms a

r = 0.05;
solve(a == (r+1) * a - 2000, a)
```

```octave
graphics_toolkit("gnuplot");

time = 25; %  Years
widthdraw = 2000; % USD

n = 0:1:time;
an = zeros(3, time+1);
r = 0.05;
an(1, 1) = 40000;
an(2, 1) = 45000;
an(3, 1) = 35000;

for i = 1:1:time,
    an(1, i+1) = (r+1) * an(1, i) - widthdraw;
    an(2, i+1) = (r+1) * an(2, i) - widthdraw;
    an(3, i+1) = (r+1) * an(3, i) - widthdraw;
end

% a(n+1) = (r+1)*a(n)

plot(n, an(1,:), "b", n, an(2,:), "r", n, an(3,:), "g");
xlabel("Time (Years - n)");               % Label for x-axis
ylabel("Account Money [USD - a(n)]");            % Label for y-axis
title("Plot of Account Money vs Time"); % Title of the graph
legend("a0 = 40000", "a0 = 45000", "a0 = 35000");
grid on;   

printf("Deposit when a0 = 40000 is: %.2f\n", an(1, time+1)); % 25th year
printf("Deposit when a0 = 45000 is: %.2f\n", an(2, time+1)); % 25th year
printf("Deposit when a0 = 35000 is: %.2f\n", an(3, time+1)); % 25th year
```

```octave
pkg load symbolic

syms a

r = 0.05;
result = solve(a == (r+1) * a - 2000, a);

% Convert symbolic result to numeric
result_numeric = double(result);

printf("Observation:\n");
printf("The equilibrium value is approximately %.2f. All curves stabilize around this value.\n", result_numeric);
```

### Stable vs. Unstable Equilibrium

#### 1. Stable Equilibrium Point
An equilibrium is considered **stable (attracting)** if small deviations from it tend to diminish over time, causing the system to return to the equilibrium.

- In this case, for any initial deposit \(a_0\), the recurrence relation brings the balance closer to ($a_{\text{eq}} = 40,000$) over time.
- The stability is a result of the fact that the multiplier \(1.05\) is close to 1, and the constant \(-2000\) acts as a damping term that pulls the balance toward equilibrium.

---

#### 2. Unstable Equilibrium Point
An equilibrium is **unstable** if small deviations from it grow over time, causing the system to move away from the equilibrium.

- For example, if the multiplier were greater than a critical value (e.g., $>2$), the balance might diverge from ($a_{\text{eq}}$). Similarly, if the withdrawal amount were lower or absent, the balance could grow indefinitely.

---

### Testing Stability: Theoretical Justification
To test whether the equilibrium point ($a_{\text{eq}}$) is stable, analyze the derivative of the recurrence relation:

1. Linearize the recurrence:
   [$
   \Delta a_{n+1} = 1.05 \Delta a_n,
   $]
   where ($\Delta a_n = a_n - a_{\text{eq}}$).

2. The multiplier \(1.05\) determines how \(\Delta a_n\) evolves:
   - If \(|1.05 - 1| < 1\), deviations diminish, and the equilibrium is stable.
   - If \(|1.05 - 1| > 1\), deviations grow, and the equilibrium is unstable.

---

### Conclusion
Here, since the multiplier is only slightly above 1 and the system is damped by the \(-2000\) term, it is **stable**.


# Population Growth and Harvesting Example

$P_{n+1} = 1.1P_n - 500$

```octave
graphics_toolkit("gnuplot");

time = 25; %  Years
widthdraw = 500; % USD

n = 0:1:time;
an = zeros(3, time+1);
r = 0.1;
an(1, 1) = 5000;
an(2, 1) = 7000;
an(3, 1) = 4000;

for i = 1:1:time,
    an(1, i+1) = (r+1) * an(1, i) - widthdraw;
    an(2, i+1) = (r+1) * an(2, i) - widthdraw;
    an(3, i+1) = (r+1) * an(3, i) - widthdraw;
end

% a(n+1) = (r+1)*a(n)

plot(n, an(1,:), "b", n, an(2,:), "r", n, an(3,:), "g");
xlabel("Time (Years - n)");               % Label for x-axis
ylabel("Fish Population [USD - a(n)]");            % Label for y-axis
title("Plot of Fish Population vs Time"); % Title of the graph
legend("a0 = 5000", "a0 = 7000", "a0 = 4000");
grid on;   

printf("Deposit when a0 = 5000 is: %.2f\n", an(1, time+1)); % 25th year
printf("Deposit when a0 = 7000 is: %.2f\n", an(2, time+1)); % 25th year
printf("Deposit when a0 = 4000 is: %.2f\n", an(3, time+1)); % 25th year
```

```octave
% Parameters
r = 1.2;          % Growth rate (unstable multiplier)
harvest = 300;    % Fixed value subtracted annually
years = 25;       % Total years to simulate
initial_values = [5000, 7000, 4000]; % Initial populations

% Initialize results matrix
populations = zeros(years + 1, length(initial_values));

% Simulate population dynamics for each initial value
for i = 1:length(initial_values)
    populations(1, i) = initial_values(i); % Set initial population
    for n = 1:years
        populations(n + 1, i) = r * populations(n, i) - harvest;
    end
end

% Plot results
figure;
hold on;
for i = 1:length(initial_values)
    plot(0:years, populations(:, i), 'LineWidth', 1.5, 'DisplayName', sprintf('P0 = %d', initial_values(i)));
end
xlabel('Years');
ylabel('Population');
title('Unstable Population Dynamics');
legend('Location', 'best');
grid on;
hold off;
```
